```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
modelo_logistico = LogisticRegression()
```

```python
## Ejemplo sencillo 

## Dataset de prueba [Iris]

from sklearn import datasets
import pandas as pd
import numpy as np

iris = datasets.load_iris()
iris_df = pd.DataFrame(data= np.c_[iris['data'], iris['target'].astype(int)],
                        columns= iris['feature_names'] + ['target'])

iris_df.head()
```

```python
iris_df['target'].value_counts()
```

```python
variables_entrada=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']
variable_salida='target'

X=iris_df[variables_entrada]
y=iris_df[variable_salida]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
modelo_logistico.coef_
```

```python
## Entrenamos el modelo

modelo_logistico.fit(X_train, y_train)
```

```python
modelo_logistico.coef_
```

```python
modelo_logistico.predict(X_test)
```

```python
modelo_logistico.predict_proba(X_test)
```
